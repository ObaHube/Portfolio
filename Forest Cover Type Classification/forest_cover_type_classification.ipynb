# Portfolio Project: Forest Cover Type

![](https://extension.usu.edu/planthealth/images/factsheets/labels-archive/aspens.jpg)

This project aims to predict forest cover type (the most common kind of tree cover) based only on cartographic variables, using a Deep Learning model. 

The actual forest cover type for a given 30 x 30 meter cell was determined from US Forest Service (USFS) Region 2 Resource Information System data. The covertypes (i.e: labels) are the following:

- Spruce/Fir
- Lodgepole Pine
- Ponderosa Pine
- Cottonwood/Willow
- Aspen
- Douglas-fir
- Krummholz

Independent variables were then derived from data obtained from the US Geological Survey and USFS. 

This study area includes four wilderness areas located in the Roosevelt National Forest of northern Colorado. These areas represent forests with minimal human-caused disturbances, so existing forest cover types are mainly a result of ecological processes rather than forest management practices.

We can now preprocess the data, with `StandardScaler`.

```python
from sklearn.base import BaseEstimator, TransformerMixin

class CustomScaler(BaseEstimator,TransformerMixin): 
    def __init__(self,columns):
        self.scaler = StandardScaler()
        self.columns = columns

    def fit(self, X, y=None):
        self.scaler.fit(X[self.columns], y)
        return self

    def transform(self, X, y=None, copy=None):
        init_col_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), columns=self.columns)
        X_not_scaled = X.iloc[:,~X.columns.isin(self.columns)]
        return pd.concat([X_not_scaled, X_scaled], axis=1)[init_col_order]
```

Finally, we split the data in train and test set.

## Step 2: Building the model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import  InputLayer
from tensorflow.keras.layers import  Dense
from sklearn.metrics import classification_report
```

### 1<sup>st</sup> attempt: two hidden layers with 20 and 10 units and `relu` activation

```python
#design the model
model = Sequential()
#add the input layer
model.add(InputLayer(input_shape=(X_train.shape[1],)))
#add a hidden layer
model.add(Dense(20, activation='relu'))
#add a hidden layer
model.add(Dense(10, activation='relu'))
#add an output layer
model.add(Dense(7, activation='softmax'))

print(model.summary())
```

We fit the model with 10 `epochs` and 64 `batch_size` to make a first attempt.

```python
#compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

#train and evaluate the model
history = model.fit(X_train, y_train, epochs = 10, batch_size = 64, verbose = 1, validation_data=(X_test, y_test))

#get additional statistics
y_estimate = model.predict(X_test)
y_estimate = np.argmax(y_estimate, axis = 1)
y_true = np.argmax(y_test, axis = 1)
print(classification_report(y_true, y_estimate))
```

```python
print(history.history.keys())
```

```python
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, 'b', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
plt.plot(epochs, acc, 'b', label='Training acc')
plt.plot(epochs, val_acc, 'r', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show
```

The model results in a general accuracy of 79%. Looking at the F1-score, it seems that there are poor performances with classes 3 (Cottonwood/Willow), 4 (Aspen), 5 (Douglas-fir), probably due to the small number of samples available in the training data.

### 2<sup>nd</sup> attempt: three hidden layers with 128, 64 and 32 units and `relu` activation

The values of `epochs` and `batch_size` are the same as the previous attempt, to check whether changing the status of hidden layers would result in a better outcome.

```python
#design the model
model = Sequential()
#add the input layer
model.add(Dense(128, activation='relu', input_shape=(X_train.shape[1],)))
#add a hidden layer
model.add(Dense(64, activation='relu'))
#add a hidden layer
model.add(Dense(32, activation='relu'))
#add an output layer
model.add(Dense(7, activation='softmax'))

#compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

#train and evaluate the model
history = model.fit(X_train, y_train, epochs = 10, batch_size = 64, verbose = 1, validation_data=(X_test, y_test))

#get additional statistics
y_estimate = model.predict(X_test)
y_estimate = np.argmax(y_estimate, axis = 1)
y_true = np.argmax(y_test, axis = 1)
print(classification_report(y_true, y_estimate))
```

```python
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, 'b', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
plt.plot(epochs, acc, 'b', label='Training acc')
plt.plot(epochs, val_acc, 'r', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show
```

With three hidden layers the outcome is indeed better, with an increase of 10% of accuracy. Also the classes 3, 4, 5 show better F1-score, even if they are still under-rated compared to the other classes.

### Hyperparameter tuning

A `GridSearch` can help us tuning the hyperparameters. These are the ones analyzed:
- `optimizer`: adam or sgd
- `activation`: relu or tanh
- \# of `units`: 32, 64 or 128
- \# of `hidden_layers`: 2, 3 or 4
- `batch_size`: 8, 32, 64 or 128
- \# of `epochs`: 10, 50 or 100

```python
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier

# Define a function to create the model
def create_model(optimizer='adam', activation='relu', units=64, hidden_layers=3):
    model = Sequential()
    model.add(Dense(units, activation=activation, input_shape=(X_train.shape[1],)))
    for _ in range(hidden_layers - 1):
        units = units/2
        model.add(Dense(units, activation=activation))
    model.add(Dense(7, activation='softmax'))
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Create a KerasClassifier wrapper for use in GridSearchCV
model = KerasClassifier(build_fn=create_model, verbose=0)

# Define the hyperparameters grid to search
param_grid = {
    'optimizer': ['adam', 'sgd'],
    'activation': ['relu', 'tanh'],
    'units': [32, 64, 128],
    'hidden_layers': [2, 3, 4],
    'batch_size': [8, 32, 64, 128], 
    'nb_epoch': [10, 50, 100]
}

# Perform GridSearchCV
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=3)
grid_search.fit(X_train, y_train)

# Print the best parameters and score
print("Best parameters found: ", grid_search.best_params_)
print("Best score found: ", grid_search.best_score_)
```

### 3<sup>rd</sup> attempt: Hyperparameters found with GridSearchCV

The last attemps uses the best hyperparameters found in the `GridSearch` 

```python
#design the model
model2 = Sequential()
#add the input layer
model2.add(Dense(128, activation='relu', input_shape=(X_train.shape[1],)))
#add a hidden layer
model2.add(Dense(64, activation='relu'))
#add a hidden layer
model2.add(Dense(32, activation='relu'))
#add an output layer
model2.add(Dense(7, activation='softmax'))

#compile the model
model2.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

#train and evaluate the model
history = model2.fit(X_train, y_train, epochs = 100, batch_size = 8, verbose = 1, validation_data=(X_test, y_test))

#get additional statistics
y_estimate = model2.predict(X_test)
y_estimate = np.argmax(y_estimate, axis = 1)
y_true = np.argmax(y_test, axis = 1)
print(classification_report(y_true, y_estimate))
```

```python
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, 'b', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
plt.plot(epochs, acc, 'b', label='Training acc')
plt.plot(epochs, val_acc, 'r', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show
```

Looking at the results, it seems that the best parameters found with `GridSearch` are not performing better than the parameters used in the 2<sup>nd</sup> attempt. 
Actually, the accuracy settles around 20 epochs. Seems thus that `GridSearch` failed in finding the best parameters.

## Step 3: Conclusions

Looking at all the results obtained, we can conclude the following:

- The best model is the one with 3 hidden layers (with 128, 64 and 32 units respectively), giving a general accuracy of 89%: that's about 75% more of random guessing (which has a likelyhood of 1/7 = 14,3%).
- With a batch size of 64 and 10 epochs, the result is very good, and also the computational time is acceptable (3 minutes).
- Looking at the different labels, we can see that the performances strongly depend on the number of samples available: poorest results are associated with classes 3 (Cottonwood/Willow), 4 (Aspen), 5 (Douglas-fir), which have very small sample sizes. Nevertheless, the f1-score of these classes is around 70%: recall seems to be highly impacted by the small sample size, while precision is always >80%.
- `GridSearch` defined a set of _optimal parameters_ which seems not optimal at all. In particular, `batch_size` and `epochs` seem not well defined: looking at the plots of _accuracy vs epochs_, we can see that we reach a peak of accuracy (about 92%) around epoch 50, and then we have a slow decrease. Considering also the average of 110 seconds per epoch needed, the set of parameters defined by `GridSearch` does not give significant improvements compared to the results obtained with the second approach.

With these results, next steps can be de following:

- Investigate how the performances can improve using _data augmentation_ techniques to improve the sample sizes of classes 3, 4 and 5.
- Investigate why `GridSearch` gave such a poor result, and try a similar approach with `RandomSearch`, in order to see whether the performances are better.
